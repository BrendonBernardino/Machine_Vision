```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
path_old = "estacionamento2.jpg"
path = "Carros-estrada.jpg"
path2 = "Images/car.jpg"
img = cv.imread(path2)
```

```python
img.shape
```

```python
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
img_resized = cv.resize(img, (800,600))
```

```python
plt.axis("off")
plt.imshow(img_resized)
plt.show()
```

```python
# img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img_resized, cv.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(img2)
plt.show()
```

```python
gray_img = cv.cvtColor(img2, cv.COLOR_RGB2GRAY)
plt.axis("off")
plt.imshow(gray_img, cmap = plt.cm.gray)
plt.show()
```

```python
path_classifier = "Cascades/cars.xml"
detector_carros = cv.CascadeClassifier(path_classifier)
```

```python
deteccoes = detector_carros.detectMultiScale(gray_img, scaleFactor=1.015, minNeighbors=5, minSize=(31,31))
# deteccoes
len(deteccoes)
```

```python
# img_detect = img
for x, y, w, h in deteccoes:
    print(w, h)
    cv.rectangle(img2, (x,y), (x+w,y+h), (0,255,255), 3)
plt.axis("off")
plt.imshow(img2)
plt.show()
```
